# 조건문

- if문

주어진 조건이 True일 경우에만 특정 코드 실행

- elif문

첫 번째 if 조건이 거짓이고, 다른 조건을 평가하고자 할 때 실행

- else문

if 문의 조건이 False일 경우에 실행

“ : ” - 구문 시작 표시
* 들여쓰기에 따라 같이 실행될지 안될지 결정됨.

```python
score = int(input("점수를 입력해주세요 :"))
if score >= 90:
	print("A등급")
elif score >= 80:
	print("B등급")
else:
	print("C등급")
```

## while 반복문

주어진 조건이 참인 동안 코드 반복 실행 * 주의사항 : 무한루프 되지않도록 코드 신경써주기

- break문 - while 문을 조건과 관계없이 강제로 탈출하는 명령어
- continue문 - continue 아래의 문장을 무시하고 while문의 다음 루프 실행

```python
num = 0
while num < 3:
	num += 1
	print("Hello, World!")
# Hello, World!
# Hello, World!
# Hello, World!
```

## for 반복문

주어진 시퀀스의 요소에 대해 코드를 반복 실행

- range(번호) - 0부터 번호미만까지 1씩 증가하며 반복
- range(시작 숫자, 번호, 간격) - 시작숫자부터 간격만큼 더해가며, 번호에 도달하기 직전까지 반복, 간격은 음수도 가능

```python
for i in range(1,5):
    print(i)
# 1
# 2
# 3
# 4
```

## 함수

함수 선언 - def 함수이름 (매개변수):
실행할 코드
return 반환값

함수 호출 - 함수 내부 코드를 실행하고 결과를 반환

함수 결과 - 반환값 미설정 시 none 반환, 반환값 여러개도 가능(여러개 반환 시 튜플 형식으로 반환)

함수 안에서 선언한 변수는 지역 변수로 함수 바깥에서 사용 불가

함수 밖에서 선언한 변수는 전역변수로 함수 안에서 사용 가능

함수의 장점
- 코드 재사용성을 높여 개발시간 단축
- 코드 가독성과 유지보수성 향상
- 코드 복잡성을 줄이고 관리가 용이하게 함

함수의 단점
- 함수의 메서드가 너무 많아지면 관리가 어려움
- 함수와 메서드의 오류는 전체 프로그램에 영향을 줌
- 복잡한 함수는 이해가 어렵고 디버깅이 어려움

```python
def add(a, b):
	c = a + b
	return c

num1 = 10
num2 = 20
result = add(num1, num2) 
print(result)
```

### 메서드(method) - 특정 객체에서 호출 되며, 그 객체의 상태를 변경하거나 동작을 수행
- 특정 개체에 종속되어 사용된다
- 객체 이름 뒤에 점(.)을 붙이고 메서드 이름을 써서 호출
- ex) list.append('a')

### 모듈(module) - 관련된 함수, 클래스, 변수 등을 하나의 파일로 모아 놓은 것
- 모듈은 일반적인 파이썬 파일
- 독립적인 모듈은 다른 모듈에 영향을 미치지 않고 개발 및 수정이 가능
- 파이썬 파일에서 import 키워드로 모듈 불러오기 가능

### 패키지(package) - 관련된 모듈을 하나의 디렉터리로 모아 놓은 것
- 패키지 디렉터리에는 init.py 파일이 포함되어 있어야 하며, 이 파일은 해당 디렉터리가 패키지임을 알려줌
- 패키지를 사용하면 코드의 구조화와 조직화가 용이
- 모듈 간의 이름 충돌을 방지

### 파이썬 계급도

- 변수/메서드 : 객체의 데이터와 기능
- 클래스(Class) : 객체를 만드는 틀(메서드의 집합)
- 모듈(Module) : .py 파일(클래스, 함수, 변수들을 모은 파일)
- 패키지(Package) : 모듈들을 모아놓은 폴더(init.py 포함)
- 라이브러리(Libarary) : 여러 패키지와 모듈을 모아놓은 거대한 기능 꾸러미
