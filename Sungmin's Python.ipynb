### 1.2 연산자

연산자 우선순위: 제곱 연산자 -> 곱셈, 나눗셈, 나머지, 몫 연산자 -> 덧셈, 뺄셈 연산자

괄호를 사용해 연산자 우선순위 변경 가능.

**덧셈**: +

**뺄셈**: -

**곱셈**: *

**나눗셈**: /

**제곱**: **

**나머지**: %

**몫**: //

**복합 연산자**: +=, -=, *=, /=, //=, %=, **=

### 1.3 문자열 자료형(str)
- 큰 따옴표: " " 

"안녕하세요 제 이름은 송성민입니다"<br><br>

- 작은 따옴표: ' '

'안녕하세요 제 이름은 송성민입니다'<br><br>

- 큰 따옴표 3개: """ """

- 작은 따옴표 3개: ''' '''

'''안녕하세요 제 이름은 송성민입니다'''

### 1.4 이스케이프 문자

\로 시작되는 문자열로 줄바꿈, 탭, 따옴표 등을 표현하기 위해 사용함.
- \n, 줄 바꿈 문자
- \t, 탭 문자
- \\\\, 백슬래쉬 문자
- \\', 작은따옴표 문자
- \\", 큰따옴표 문자


### 1.5 문자열 연산
문자열끼리 연산이 가능하다.
- 문자열 덧셈: 문자열 + 문자열
- 문자열 곱셈: 문자열 * 숫자

```python
a = "Hello "
b = "World"

# 문자열 덧셈 "Hello " + "World" = "Hello World"
text = a + b
print(text)

# 문자열 곱셈 "Hello " * 5 = "Hello Hello Hello Hello Hello"
text = a * 5
print(text)
```
