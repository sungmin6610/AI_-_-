## 자료형
### 1.1 숫자형 자료형
**int**
: 정수를 뜻하는 자료형
(ex) 0, -1, 1

**float**
: 실수를 뜻하는 자료형
(ex) 0.0, 0.1, -0.1

```python
# 변수 선언 : 변수이름 = 값
# 변수가 자료형을 스스로 판단하여 지정해준다.
# type() 함수를 사용하여 변수의 자료형을 확인할 수 있다.

a = 10      # 변수 a에 정수 10의 값을 할당한다.
b = 10.123  # 변수 b에 실수 10.123의 값을 할당한다.

print(a)
print(b)

print(type(a))    # 변수 a의 자료형을 확인하고, 출력한다.
print(type(b))
```

### 1.2 연산자

연산자 우선순위: 제곱 연산자 -> 곱셈, 나눗셈, 나머지, 몫 연산자 -> 덧셈, 뺄셈 연산자

괄호를 사용해 연산자 우선순위 변경 가능.

**덧셈**: +

**뺄셈**: -

**곱셈**: *

**나눗셈**: /

**제곱**: **

**나머지**: %

**몫**: //

**복합 연산자**: +=, -=, *=, /=, //=, %=, **=

```python
a = 10
b = 5

result=a+b      # a(10) + b(5)의 값을 result에 저장한다.
print(result)

result = a - b      # a(10) - b(5)의 값을 result에 저장한다.
print(result)

result = a * b      # a(10) X b(5)의 값을 result에 저장한다.
print(result)

# 다른 연산자는 결과 값이 동일한 자료형으로 나오는데,
# 나머지 연산자는 무조건, 실수형으로 나옵니다.
result = a / b      # a(10) / b(5)의 값을 result에 저장한다.
print(result)

result = a % b      # a(10) % b(5)의 값을 result에 저장한다.
print(result)       # 10 / 5 = 몫 : 2, 나머지 : 0
print(10 % 7)       # 10 / 7 = 몫 : 1, 나머지 : 3

result = a**b       # a(10) ** b(5)의 값을 result에 저장한다.
print(result)

result = a//b       # a(10) // b(5)의 값을 result에 저장한다.
print(result)       # 10 / 5 = 몫 : 2, 나머지 : 0

result = 2 + 3 * 4    # 14
print(result)

result = (2 + 3) * 4  # 20
print(result)
```

### 1.3 문자열 자료형(str)
- 큰 따옴표: " " 

"안녕하세요 제 이름은 송성민입니다"<br><br>

- 작은 따옴표: ' '

'안녕하세요 제 이름은 송성민입니다'<br><br>

- 큰 따옴표 3개: """ """

"""안녕하세요 제 이름은 송성민입니다"""<br><br>

- 작은 따옴표 3개: ''' '''

'''안녕하세요 제 이름은 송성민입니다'''

```python
# 문자열 안에 ""를 사용하기 위해 ''로 문자열을 구성.
a = 'a, 누구는 말했다."최선을 다하라"\n'
print(a)

# 문자열 안에 ''를 사용하기 위해 ""로 문자열을 구성.
b = "b, 누구는 말했다. '최선을 다하라'\n"
print(b)

# ''' '''로 씌여진 문자열은 줄바꿈, '', "" 등 거의 모든 형식을 표현할 수 있다.
c = '''c, "안"녕
하'세'
요\n'''
print(c)

# """ """로 씌여진 문자열은 줄바꿈, '', "" 등 거의 모든 형식을 표현할 수 있다.
d = """d, 안녕하
세요"""
print(d)
```

### 1.4 이스케이프 문자

\로 시작되는 문자열로 줄바꿈, 탭, 따옴표 등을 표현하기 위해 사용함.
- \n, 줄 바꿈 문자
- \t, 탭 문자
- \\\\, 백슬래쉬 문자
- \\', 작은따옴표 문자
- \\", 큰따옴표 문자


```python
# \' 작은따옴표 문자를 이용하여 ''로 구성된 문자열 안에서 '사용
a = '누구는 말했다.\n\'최선을 다하라\''
print(a)

# \t 탭 문자를 사용하여 길이가 다른 두 문자열을 일정한 길이로 유지.
b = "115\t23\t4516"
c = "1\t31255\t456"
print(b)
print(c)
```

### 1.5 문자열 연산
문자열끼리 연산이 가능하다.
- 문자열 덧셈: 문자열 + 문자열
- 문자열 곱셈: 문자열 * 숫자

```python
a = "Hello "
b = "World"

# 문자열 덧셈 "Hello " + "World" = "Hello World"
text = a + b
print(text)

# 문자열 곱셈 "Hello " * 5 = "Hello Hello Hello Hello Hello"
text = a * 5
print(text)
```
