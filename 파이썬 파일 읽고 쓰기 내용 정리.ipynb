# 파일 읽고 쓰기

- 파일이 없으면 파일을 생성하면서 열고, 파일이 있으면 해당 파일을 연다.
- open()을 썼으면 close()도 무조건 써줘야함, open, close는 세트
- 'r' (읽기 모드): 파일을 읽기만 할 때 사용하며, 파일이 없으면 에러가 발생함
- 'w' (쓰기 모드): 파일에 내용을 쓸 때 사용. 파일이 없으면 새로 생성하고, 이미 파일이 있으면 기존 내용을 모두 지우고 새로 씀.
- 'a' (추가 모드): 파일의 마지막에 새로운 내용을 추가할 때 사용. 기존 내용을 보존하면서 내용을 덧붙일 수 있음.

```python
f = open("filereadwrite.txt", 'w')
f.close()
```

- 파일은 생성할 수 있지만 폴더 생성은 못함

```python
f = open("C:/0326/filereadwrite.txt", 'w')
f.close()
```

## 파일을 쓰기 모드로 열어 내용 쓰기

f.write()로 파일에 내용을 넣어줄 수 있다.

```python
f = open("filereadwrite.txt", 'w')

for i in range(1, 11):
    data = "%d번째 줄입니다.\n" % i
    print(data)
    f.write(data)

print("문자열 파일에 쓰기")
f.write("문자열 파일에 쓰기")
f.close()
```

## readlines 함수

파일의 모든 줄을 읽어서 각각의 줄을 요소로 갖는 리스트로 반환한다

```python
f = open("filereadwrite.txt", 'r')
lines = f.readlines()
print(lines)
f.close()
```

```python
f = open("filereadwrite.txt", 'r')
lines = f.readlines()
for line in lines:
    print(line)
f.close()
```

## read 함수

파일의 내용 전체를 하나의 문자열로 읽어온다.

```python
f = open("filereadwrite.txt", 'r')
data = f.read()
print(data)
f.close()
```

## 파일 객체를 for 문과 사용하기

파일 객체를 직접 for 문에 넣어 사용하면 한 줄씩 순차적으로 읽어와 메모리 효율이 좋다.

```python
f = open("filereadwrite.txt", 'r')
for line in f:
    print(line)
f.close()
```

## 파일에 새로운 내용 추가

- 'a' 모드를 사용하면 기존 내용 뒤에 내용 추가 가능

```python
f = open("filereadwrite.txt",'a')
f.write('\n')
for i in range(11, 20):
    data = "%d번째 줄입니다.\n" % i
    f.write(data)
f.close()
```

## with 문 사용
- with 문 활용: with를 사용하면 블록이 끝나는 순간 파일이 자동으로 닫히므로 close()를 따로 쓰지 않아도 된다.

```python
with open("difficult.txt", 'w') as f:
    f. write("I like Python. But it is difficult")
```

파일 처리 시 주의사항

- 한글이 포함된 파일을 다룰 때는 인코딩 명시

```python
#한글 파일 쓰기
with open("한글파일.txt", "w", encoding="utf-8") as f:
    f.write("안녕하세요, 파이썬!")

# 한글 파일 읽기
with open("한글파일.txt", "r", encoding="utf-8") as f:
    content = f.read()
    print(content)
```
